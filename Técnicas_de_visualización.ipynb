```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.DataFrame({
    'Categoría': ['Electrónica', 'Ropa', 'Hogar'],
    'Ventas': [200, 150, 300]
})

# Gráfico de barras
data.plot(kind='bar', x='Categoría', y='Ventas')
plt.title('Ventas por Categoría')
plt.show()
```
