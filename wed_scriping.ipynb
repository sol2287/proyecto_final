```python
# importar paquetes
from bs4 import BeautifulSoup
import requests
```

```python
#linq del que extraeremos la data
lapagina = 'https://subslikescript.com/movie/Dragons__Damsels-27262154'
#para enviar una solicitud
respuesta = requests.get(lapagina)
#obtener el texto del resultado de la pagina despues del requests
contenido = respuesta.text
#crea variables soup que nos permite localizar elementos en la pagina wed
soup = BeautifulSoup (contenido,'html')
print(soup.prettify())#la funcion prettify despliega el html ordenado
```

```python
import os
from google.colab import files
referencia =soup.find('article', class_='main-article')
titulo = referencia.find ('h1').get_text()
titulillo = soup.find('h1').get_text()
print('el titulo de la pagina es: ', titulo)
palabras = referencia.find('div').get_text(strip=True, separator=' ')
print('las palabras son : ', palabras)
with open('mi_sol12.txt', 'w') as file:
  file.write(palabras)
print("Directorio axtual de trabajo:", os.getcwd())
print("Archivos en el dirctorio actual:", os.listdir())
files.download('mi_sol12.txt')
```
